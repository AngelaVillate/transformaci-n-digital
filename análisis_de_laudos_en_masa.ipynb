```python
# Cargar modelo de spaCy en español
nlp = spacy.load("es_core_news_sm")
```

```python
# prompt: descomprime carpeta, lee cada uno de los archivos pdf y muestra el texto de los pdf /content/laudos muestra.zip

import os
import zipfile

def extract_text_from_pdfs(zip_path):
  """
  Extrae el texto de todos los archivos PDF dentro de un archivo ZIP.

  Args:
    zip_path: Ruta al archivo ZIP.
  """
  with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('extracted_laudos')

  for root, _, files in os.walk('extracted_laudos'):
    for file in files:
      if file.endswith('.pdf'):
        pdf_path = os.path.join(root, file)
        try:
          with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages)
            text = ''
            for page_num in range(num_pages):
              page = pdf_reader.pages[page_num]
              text += page.extract_text()
            print(f"Archivo PDF: {file}\nTexto extraído:\n{text}\n{'='*30}")
        except Exception as e:
          print(f"Error al procesar el archivo {file}: {e}")


# Ruta al archivo ZIP
zip_file_path = '/content/laudos muestra.zip'

# Extraer el texto de los PDFs
extract_text_from_pdfs(zip_file_path)
```

```python
# prompt: cada archivo y su texto en un df

import os
import pandas as pd
import PyPDF2
import zipfile

def extract_text_from_pdfs_to_df(zip_path):
  """
  Extrae el texto de todos los archivos PDF dentro de un archivo ZIP y lo guarda en un DataFrame.

  Args:
    zip_path: Ruta al archivo ZIP.
  """
  with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('extracted_laudos')

  data = []
  for root, _, files in os.walk('extracted_laudos'):
    for file in files:
      if file.endswith('.pdf'):
        pdf_path = os.path.join(root, file)
        try:
          with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages)
            text = ''
            for page_num in range(num_pages):
              page = pdf_reader.pages[page_num]
              text += page.extract_text()
            data.append({'Archivo': file, 'Texto': text})
        except Exception as e:
          print(f"Error al procesar el archivo {file}: {e}")

  return pd.DataFrame(data)


# Ruta al archivo ZIP
zip_file_path = '/content/laudos muestra.zip'

# Extraer el texto de los PDFs y guardar en un DataFrame
df_laudos = extract_text_from_pdfs_to_df(zip_file_path)

# Imprimir el DataFrame
df_laudos
```

```python
# prompt: genera nueva columna con unicode, minusculas, stop words en español

def preprocess_text(text):
  """
  Preprocesa el texto: convierte a Unicode, minúsculas, elimina stop words.
  """
  if isinstance(text, str):  # Check if text is a string
    text = text.lower()
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop and token.is_alpha]
    processed_text = " ".join(tokens)
    return processed_text
  else:
    return ""  # Return an empty string if text is not a string


# Aplicar la función de preprocesamiento a la columna 'Texto'
df_laudos['Texto_procesado'] = df_laudos['Texto'].apply(preprocess_text)

# Imprimir el DataFrame con la nueva columna
df_laudos
```

```python
# prompt: a partir de la columna Texto_procesado lematiza y tokeniza

def lemmatize_and_tokenize(text):
  """
  Lematiza y tokeniza el texto utilizando spaCy.

  Args:
    text: El texto a procesar.

  Returns:
    Una lista de tokens lematizados.
  """
  if isinstance(text, str):
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]
    return lemmas
  else:
    return []


# Aplicar la función de lematización y tokenización a la columna 'Texto_procesado'
df_laudos['Lemmas_Tokens'] = df_laudos['Texto_procesado'].apply(lemmatize_and_tokenize)

# Imprimir el DataFrame con la nueva columna
df_laudos
```

```python
# prompt: a partir de Lemas_tokens vectoriza con tfidf

# Crear una instancia de TfidfVectorizer
vectorizer = TfidfVectorizer()

# Ajustar y transformar los Lemmas_Tokens a una matriz TF-IDF
tfidf_matrix = vectorizer.fit_transform(df_laudos['Lemmas_Tokens'].apply(lambda x: ' '.join(x)))

# Imprimir la matriz TF-IDF
print(tfidf_matrix)

# Puedes acceder a los términos del vocabulario con:
print(vectorizer.get_feature_names_out())
```

```python
# prompt: a partir de la vectoización de TFIDF calcula la distancia del coseno entre cada uno de los archivos

import pandas as pd
# Calcular la distancia del coseno entre los documentos
cosine_similarities = cosine_distances(tfidf_matrix)

# Crear un DataFrame con las distancias del coseno
df_cosine_similarities = pd.DataFrame(cosine_similarities, index=df_laudos['Archivo'], columns=df_laudos['Archivo'])

# Imprimir el DataFrame de distancias del coseno
df_cosine_similarities
```

```python
# prompt: representa la similitud en un mapa de calor

import matplotlib.pyplot as plt
# Crear un mapa de calor de las distancias del coseno
plt.figure(figsize=(12, 10))
sns.heatmap(df_cosine_similarities, annot=True, cmap='viridis', fmt=".2f")
plt.title('Similitud entre Laudos (Distancia del Coseno)')
plt.show()
```

```python
# prompt: en df_laudos en la columna Lemas_tokens identifica topic modeling y visualiza

# Crear una lista de textos procesados
texts = df_laudos['Lemmas_Tokens'].tolist()

# Crear un diccionario de palabras y sus frecuencias
dictionary = corpora.Dictionary(texts)

# Crear un corpus de documentos
corpus = [dictionary.doc2bow(text) for text in texts]

# Entrenar el modelo LDA
lda_model = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, random_state=42)  # Ajustar num_topics según sea necesario

# Visualizar los tópicos
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpus, dictionary)
vis
```

```python
# prompt: en df_laudos en columnas Lemas_tokens identifica los NER en español y grafica los NER más comunes por cada archivo

import matplotlib.pyplot as plt
def identify_ner_and_plot(df_laudos):
  """
  Identifica los NER en español en la columna 'Lemmas_Tokens' y grafica los NER más comunes por cada archivo.

  Args:
    df_laudos: DataFrame con la columna 'Lemmas_Tokens'.
  """
  for index, row in df_laudos.iterrows():
    lemmas_tokens = row['Lemmas_Tokens']
    if lemmas_tokens:
      text = " ".join(lemmas_tokens)  # Convertir la lista de tokens a una cadena
      doc = nlp(text)
      ner_labels = [(ent.text, ent.label_) for ent in doc.ents]

      if ner_labels:
        ner_counts = Counter([label for _, label in ner_labels])
        print(f"\nArchivo: {row['Archivo']}")
        print("NER más comunes:")

        # Crear el gráfico de barras
        plt.figure(figsize=(10, 5))
        plt.bar(ner_counts.keys(), ner_counts.values())
        plt.title(f"NER más comunes en {row['Archivo']}")
        plt.xlabel("NER")
        plt.ylabel("Frecuencia")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        for ner, count in ner_counts.items():
          print(f"  - {ner}: {count}")
      else:
        print(f"\nArchivo: {row['Archivo']}")
        print("No se encontraron NER en este archivo.")


# Llamar a la función para identificar NER y graficar
identify_ner_and_plot(df_laudos)
```

```python
# prompt: genera una lista que relacione NER y las palabras asociadas

def generate_ner_word_list(df_laudos):
  """
  Genera una lista que relaciona NER y las palabras asociadas.

  Args:
    df_laudos: DataFrame con la columna 'Lemmas_Tokens'.

  Returns:
    Una lista de tuplas (NER, palabra).
  """
  ner_word_list = []
  for index, row in df_laudos.iterrows():
    lemmas_tokens = row['Lemmas_Tokens']
    if lemmas_tokens:
      text = " ".join(lemmas_tokens)
      doc = nlp(text)
      for ent in doc.ents:
        ner_word_list.append((ent.label_, ent.text))
  return ner_word_list


# Obtener la lista de NER y palabras
ner_word_list = generate_ner_word_list(df_laudos)

# Imprimir la lista
ner_word_list
```

```python
# prompt: genera lista de ner y palabras asociadas por cada archivo

def generate_ner_word_list_per_file(df_laudos):
  """
  Genera una lista de NER y palabras asociadas por cada archivo.

  Args:
    df_laudos: DataFrame con la columna 'Lemmas_Tokens'.

  Returns:
    Un diccionario donde la clave es el nombre del archivo y el valor es una lista de tuplas (NER, palabra).
  """
  ner_word_list_per_file = {}
  for index, row in df_laudos.iterrows():
    file_name = row['Archivo']
    lemmas_tokens = row['Lemmas_Tokens']
    if lemmas_tokens:
      text = " ".join(lemmas_tokens)
      doc = nlp(text)
      ner_word_list = []
      for ent in doc.ents:
        ner_word_list.append((ent.label_, ent.text))
      ner_word_list_per_file[file_name] = ner_word_list
  return ner_word_list_per_file


# Obtener el diccionario de NER y palabras por archivo
ner_word_list_per_file = generate_ner_word_list_per_file(df_laudos)

# Imprimir el diccionario
ner_word_list_per_file
```
